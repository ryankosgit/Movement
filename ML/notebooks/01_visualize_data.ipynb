```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Ryan B datasets
df_ryanb_pushup = pd.read_excel('../data/raw/ryanb_pushup_20reps_1762618264.xlsx')
df_ryanb_jumpingjack = pd.read_excel('../data/raw/ryanb_jumping jacks_20reps_1762618810.xlsx')  # Update filename
df_ryanb_squat = pd.read_excel('../data/raw/ryanb_squat_20reps_1762615894.xlsx')  # Update filename

# Load Ryan K datasets
df_ryank_pushup = pd.read_excel('../data/raw/ryank_pushup_20reps_1762612126.xlsx')
df_ryank_jumpingjack = pd.read_excel('../data/raw/ryank_jumping_jack_20reps_1762615339.xlsx')  # Update filename
df_ryank_squat = pd.read_excel('../data/raw/ryank_squat_20reps_1762617710.xlsx') 

# Check they loaded
print("Ryan B Pushup shape:", df_ryanb_pushup.shape)
print("Ryan B Jumping Jack shape:", df_ryanb_jumpingjack.shape)
print("Ryan B Squat shape:", df_ryanb_squat.shape)
print("Ryan K Pushup shape:", df_ryank_pushup.shape)
print("Ryan K Jumping Jack shape:", df_ryank_jumpingjack.shape)
print("Ryan K Squat shape:", df_ryank_squat.shape)
```

```python
# Plot Ryan B Pushup data
plt.figure(figsize=(15, 8))
plt.suptitle('Ryan B - Pushup Dataset', fontsize=16, fontweight='bold')

plt.subplot(3, 1, 1)
plt.plot(df_ryanb_pushup['p_ay'], label='Phone Y (Ryan B Pushup)', alpha=0.7)
plt.title('Phone Acceleration (Torso Movement)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(df_ryanb_pushup['a_ay'], label='AirPods Y (Ryan B Pushup)', alpha=0.7)
plt.title('AirPods Acceleration (Head Movement)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(df_ryanb_pushup['p_gy'], label='Phone Gyro Y (Ryan B Pushup)', alpha=0.7)
plt.title('Phone Gyroscope (Rotation)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Plot Ryan B Jumping Jack data
plt.figure(figsize=(15, 8))
plt.suptitle('Ryan B - Jumping Jack Dataset', fontsize=16, fontweight='bold')

plt.subplot(3, 1, 1)
plt.plot(df_ryanb_jumpingjack['p_ay'], label='Phone Y (Ryan B JJ)', alpha=0.7, color='green')
plt.title('Phone Acceleration (Torso Movement)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(df_ryanb_jumpingjack['a_ay'], label='AirPods Y (Ryan B JJ)', alpha=0.7, color='green')
plt.title('AirPods Acceleration (Head Movement)')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(df_ryanb_jumpingjack['p_gy'], label='Phone Gyro Y (Ryan B JJ)', alpha=0.7, color='green')
plt.title('Phone Gyroscope (Rotation)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
