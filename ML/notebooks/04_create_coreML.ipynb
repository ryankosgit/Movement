```python
import coremltools as ct
import joblib
import numpy as np

# Load the saved model
rf = joblib.load("exercise_rf.pkl")

# Convert to Core ML
mlmodel = ct.converters.sklearn.convert(
    rf,
    input_features=[
        "p_ax","p_ay","p_az",
        "p_gx","p_gy","p_gz",
        "a_ax","a_ay","a_az",
        "a_gx","a_gy","a_gz"
    ],
    output_feature_names=["label"]
)

# Save to file
mlmodel.save("ExerciseClassifier.mlmodel")
print("âœ… Saved Core ML model: ExerciseClassifier.mlmodel")
```
